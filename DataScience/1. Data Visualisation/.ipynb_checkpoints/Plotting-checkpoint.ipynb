```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
houses = pd.read_csv("data/kc_house_data.csv")
```

```python
houses.sort_values("price", inplace=True, ascending=False)
houses
```

```python
houses.head()[["sqft_living", "sqft_lot"]].plot.bar()
```

```python
salaries = pd.read_csv("data/Salaries.csv", low_memory=False)
```

```python
salaries["BasePay"] = pd.to_numeric(salaries["BasePay"], errors="coerce").fillna(0)
salaries["OvertimePay"] = pd.to_numeric(salaries["OvertimePay"], errors="coerce").fillna(0)
salaries["OtherPay"] = pd.to_numeric(salaries["OtherPay"], errors="coerce").fillna(0)
```

```python
salaries
```

```python
df = salaries[["EmployeeName", "BasePay", "OvertimePay", "OtherPay"]]
```

```python
df.set_index("EmployeeName").head(10).plot.bar(stacked=True)
```

```python
df.set_index("EmployeeName").head(10).plot.barh(stacked=True)
```

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
df.set_index("EmployeeName").head(10).plot.bar(stacked=True)
```

```python
plt.style.use('Solarize_Light2')
df.set_index("EmployeeName").head(10).plot.bar(stacked=True)
```

```python
plt.style.use('ggplot')
df.set_index("EmployeeName").head(10).plot.bar(stacked=True, legend=False)
```

## Boxplots

Boxplot can be drawn calling Series.plot.box() and DataFrame.plot.box(), or DataFrame.boxplot() to visualize the distribution of values within each column.


```python
df = pd.DataFrame(np.random.rand(10, 5), columns=["A", "B", "C", "D", "E"])
df.plot.box()
```

```python
houses[["floors", "bedrooms", "bathrooms"]].plot.box()
```

```python
titanic.plot(subplots=True)
```

```python
titanic.age.plot(figsize=(20,10), c="teal")
```
