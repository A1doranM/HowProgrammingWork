### Double Integrals

Hello class! We've seen how a single integral `∫ f(x) dx` helps us find the area under a 2D curve. Now, we're going to step up a dimension. A double integral helps us find the **volume under a 3D surface**.

**The Core Idea:**
*   Imagine a surface defined by `z = f(x, y)`. This surface could be a mountain range, a dome, or a wavy sheet floating over the xy-plane.
*   Imagine a rectangular region `R` on the floor (the xy-plane) directly beneath this surface.
*   A double integral, `∬_R f(x, y) dA`, calculates the volume of the solid region between the floor `R` and the surface `f(x, y)`.

The `dA` represents a tiny little patch of area in the region `R`. The double integral works by chopping the floor `R` into infinitely many tiny rectangles of area `dA`. For each patch, it calculates the volume of the tall, skinny column above it (which is `height × area = f(x, y) * dA`) and then adds up the volumes of all these columns.

#### Calculating Double Integrals (Fubini's Theorem)

The magic of calculating a double integral is to turn it into two single integrals, one inside the other. This is called an **iterated integral**. We can do this in two orders:

1.  `∬_R f(x, y) dA = ∫_a^b [ ∫_c^d f(x, y) dy ] dx`
2.  `∬_R f(x, y) dA = ∫_c^d [ ∫_a^b f(x, y) dx ] dy`

**How to read this:**
*   **Work from the inside out.** For the first order (`dy dx`), you first integrate `f(x, y)` with respect to `y`, treating `x` as if it were just a constant.
*   Then, you plug in the `y`-bounds (`c` and `d`).
*   The result will be an expression that only has `x` in it. You then integrate this new expression with respect to `x` using the `x`-bounds (`a` and `b`).

---
### Detailed Calculation Example: Rectangular Region

Let's find the volume under the surface **`f(x, y) = x² + y`** over the rectangular region `R` defined by **`0 ≤ x ≤ 2`** and **`0 ≤ y ≤ 1`**.

**1. Set up the Iterated Integral**
Let's use the `dy dx` order. The outer bounds are for `x` (0 to 2), and the inner bounds are for `y` (0 to 1).
`Volume = ∫_0^2 ∫_0^1 (x² + y) dy dx`

**2. Solve the Inner Integral**
We integrate `(x² + y)` with respect to `y`, treating `x²` as a constant.
`∫_0^1 (x² + y) dy = [ x²y + y²/2 ]` evaluated from `y=0` to `y=1`.
Plug in the `y` bounds:
`= [ x²(1) + 1²/2 ] - [ x²(0) + 0²/2 ]`
`= [ x² + 1/2 ] - [ 0 ] = x² + 1/2`

**3. Solve the Outer Integral**
Now we take the result from Step 2 and integrate it with respect to `x`.
`∫_0^2 (x² + 1/2) dx = [ x³/3 + x/2 ]` evaluated from `x=0` to `x=2`.
Plug in the `x` bounds:
`= [ 2³/3 + 2/2 ] - [ 0³/3 + 0/2 ]`
`= [ 8/3 + 1 ] - [ 0 ] = 8/3 + 3/3 = 11/3`

**Result:** The volume is **11/3** cubic units.

### Finding Bounds and Double Integrals with Variable Bounds

Most regions aren't simple rectangles. They are triangles, circles, or other curvy shapes. In these cases, the inner bounds of integration will be functions, not constants.

**The Key Skill: Finding the Bounds**
Imagine the region `D` on the xy-plane.

*   **For `dy dx` order (Vertical Slices):**
    1.  Imagine the "shadow" of the region cast onto the **x-axis**. This gives you the outer, constant bounds for `x`, from `x=a` to `x=b`.
    2.  Now, imagine a vertical arrow shooting up through the region at some `x` value. The `y`-value where the arrow **enters** the region is the lower bound, `y = g₁(x)`. The `y`-value where it **exits** is the upper bound, `y = g₂(x)`.

*   **For `dx dy` order (Horizontal Slices):**
    1.  Imagine the "shadow" of the region cast onto the **y-axis**. This gives you the outer, constant bounds for `y`, from `y=c` to `y=d`.
    2.  Now, imagine a horizontal arrow shooting from left-to-right through the region at some `y` value. The `x`-value where the arrow **enters** is the lower bound, `x = h₁(y)`. The `x`-value where it **exits** is the upper bound, `x = h₂(y)`.

### Detailed Example: Switching Bounds on a Triangular Region

Let's integrate the same function **`f(x, y) = x² + y`** over a triangular region `D` bounded by the lines **`y = 0`**, **`x = 2`**, and **`y = x/2`**.

#### Calculation 1: `dy dx` Order (Vertical Slices)

**1. Find the Bounds**
*   **Outer (x):** The region's shadow on the x-axis goes from `x=0` to `x=2`. So, `0 ≤ x ≤ 2`.
*   **Inner (y):** A vertical arrow enters the region at the bottom line, `y=0`. It exits at the top diagonal line, `y=x/2`. So, `0 ≤ y ≤ x/2`.

**2. Set up and Solve**
`∫_0^2 ∫_0^(x/2) (x² + y) dy dx`
*   **Inner integral (w.r.t. y):**
    `∫_0^(x/2) (x² + y) dy = [ x²y + y²/2 ]` from `y=0` to `y=x/2`.
    `= [ x²(x/2) + (x/2)²/2 ] - [0] = x³/2 + (x²/4)/2 = x³/2 + x²/8`
*   **Outer integral (w.r.t. x):**
    `∫_0^2 (x³/2 + x²/8) dx = [ x⁴/8 + x³/24 ]` from `x=0` to `x=2`.
    `= [ 2⁴/8 + 2³/24 ] - [0] = [ 16/8 + 8/24 ] = 2 + 1/3 = 7/3`

**Result 1: 7/3**

#### Calculation 2: `dx dy` Order (Switching the Bounds)

**1. Find the Bounds**
*   **Outer (y):** The region's shadow on the y-axis goes from `y=0` up to the point where `x=2` and `y=x/2` intersect. This is `y=2/2=1`. So, `0 ≤ y ≤ 1`.
*   **Inner (x):** A horizontal arrow enters the region at the diagonal line. We must rewrite this line as `x` in terms of `y`: `y = x/2` becomes `x = 2y`. The arrow exits at the vertical line, `x = 2`. So, `2y ≤ x ≤ 2`.

**2. Set up and Solve**
`∫_0^1 ∫_(2y)^2 (x² + y) dx dy`
*   **Inner integral (w.r.t. x):**
    `∫_(2y)^2 (x² + y) dx = [ x³/3 + yx ]` from `x=2y` to `x=2`.
    `= [ (2³/3 + y(2)) ] - [ ((2y)³/3 + y(2y)) ]`
    `= [ 8/3 + 2y ] - [ 8y³/3 + 2y² ] = 8/3 + 2y - 8y³/3 - 2y²`
*   **Outer integral (w.r.t. y):**
    `∫_0^1 (8/3 + 2y - 2y² - 8y³/3) dy = [ 8y/3 + y² - 2y³/3 - 8y⁴/12 ]` from `y=0` to `y=1`.
    `= [ 8/3 + 1 - 2/3 - 8/12 ] - [0]`
    `= [ 6/3 + 1 - 2/3 ] = 2 + 1 - 2/3 = 3 - 2/3 = 9/3 - 2/3 = 7/3`

**Result 2: 7/3**

Both orders give the same answer! Sometimes one order of integration is much simpler than the other, so knowing how to switch them is a very powerful skill.

### Python Code Illustration

The Python code will calculate the integral over the triangular region using both orders of integration to verify they are equal. It will also visualize the region of integration, which is key to understanding where the bounds came from.




```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbols and the function
x, y = sp.symbols('x y')
f = x**2 + y

print(f"Integrating the function f(x, y) = {f}")
print("Over the triangular region bounded by y=0, x=2, and y=x/2.\n")

# --- Calculation 1: dy dx order ---
print("--- Method 1: dy dx order (Vertical Slices) ---")
# Inner integral with respect to y, with variable bounds
inner_integral_dy = sp.integrate(f, (y, 0, x/2))
# Outer integral with respect to x
result_dy_dx = sp.integrate(inner_integral_dy, (x, 0, 2))

print(f"Setup: ∫ from 0 to 2 [ ∫ from 0 to x/2 ({f}) dy ] dx")
print(f"Result after inner integral: {inner_integral_dy}")
print(f"Final Result = {result_dy_dx}\n")


# --- Calculation 2: dx dy order ---
print("--- Method 2: dx dy order (Horizontal Slices) ---")
# First, we need to solve y = x/2 for x -> x = 2y
# Inner integral with respect to x, with variable bounds
inner_integral_dx = sp.integrate(f, (x, 2*y, 2))
# Outer integral with respect to y
result_dx_dy = sp.integrate(inner_integral_dx, (y, 0, 1))

print(f"Setup: ∫ from 0 to 1 [ ∫ from 2y to 2 ({f}) dx ] dy")
print(f"Result after inner integral: {inner_integral_dx}")
print(f"Final Result = {result_dx_dy}\n")

print(f"Conclusion: Both orders yield {result_dy_dx}, confirming our manual calculation.")


# --- Visualization of the Region ---
fig, ax = plt.subplots(figsize=(8, 6))

# Define the x values for plotting
x_vals = np.linspace(0, 3, 400)
# Define the boundary lines
y1 = np.zeros_like(x_vals) # y = 0
y2 = x_vals / 2.0         # y = x/2

# Plot the boundaries
ax.plot(x_vals, y1, 'b-', label='y = 0')
ax.plot(x_vals, y2, 'r-', label='y = x/2')
ax.axvline(x=2, color='g', linestyle='-', label='x = 2')

# Shade the region of integration
# We need to find the interval where y2 is below the x=2 line
x_fill = np.linspace(0, 2, 200)
y_fill = x_fill / 2.0
ax.fill_between(x_fill, 0, y_fill, color='gray', alpha=0.5, label='Region of Integration D')

ax.set_title('Region of Integration D', fontsize=16)
ax.set_xlabel('x-axis', fontsize=12)
ax.set_ylabel('y-axis', fontsize=12)
ax.set_xlim(-0.5, 3)
ax.set_ylim(-0.5, 2)
ax.grid(True)
ax.legend()
plt.show()

```
