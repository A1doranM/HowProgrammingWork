# Approximating areas with Riemann sums

![](./images/img.png)

![](./images/img_1.png)

![](./images/img_2.png)

![](./images/img_3.png)

![](./images/img_4.png)

```python
import numpy as np
import matplotlib.pyplot as plt

# Function
def f(x):
    return x**2

# Interval and subintervals
a, b = 0, 2
n = 4
x = np.linspace(a, b, n+1)  # Partition points
dx = (b - a) / n  # Width of each rectangle

# Left Riemann Sum
x_left = x[:-1]
L_sum = sum(f(x_left) * dx)

# Right Riemann Sum
x_right = x[1:]
R_sum = sum(f(x_right) * dx)

# Midpoint Riemann Sum
x_mid = (x[:-1] + x[1:]) / 2
M_sum = sum(f(x_mid) * dx)

# Print results
print(f"Left Riemann Sum: {L_sum}")
print(f"Right Riemann Sum: {R_sum}")
print(f"Midpoint Riemann Sum: {M_sum}")

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(a, b, 100), f(np.linspace(a, b, 100)), 'b', label="f(x) = x^2")
plt.bar(x_left, f(x_left), width=dx, alpha=0.4, align='edge', label="Left Riemann")
plt.bar(x_right, f(x_right), width=-dx, alpha=0.4, align='edge', label="Right Riemann")
plt.bar(x_mid, f(x_mid), width=dx, alpha=0.4, label="Midpoint Riemann")
plt.legend()
plt.title("Riemann Sums Approximation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

## Riemann Trapezoidal Approximation

![](./images/img_5.png)

![](./images/img_6.png)

![](./images/img_7.png)

```python
import numpy as np

# Define the function
def f(x):
    return x**2 + 1

# Trapezoidal Rule
def trapezoidal_rule(f, a, b, n):
    x = np.linspace(a, b, n + 1)  # Divide interval into n subintervals
    y = f(x)
    dx = (b - a) / n
    area = dx * (0.5 * y[0] + 0.5 * y[-1] + sum(y[1:-1]))
    return area

# Parameters
a = 0  # Lower limit
b = 4  # Upper limit
n = 4  # Number of subintervals

# Compute the approximate integral
result = trapezoidal_rule(f, a, b, n)
print(f"Approximate value of the integral: {result}")
```

## Definite Integral as the Limit of a Riemann Sum

![](./images/img_8.png)

![](./images/img_9.png)

![](./images/img_10.png)

![](./images/img_11.png)

![](./images/img_12.png)

```python
import numpy as np

# Function to integrate
def f(x):
    return x**2

# Interval [a, b]
a = 0
b = 1

# Number of subdivisions
n = 1000
dx = (b - a) / n

# Riemann sum using right endpoints
x = np.linspace(a + dx, b, n)  # Right endpoints
riemann_sum = np.sum(f(x) * dx)

print(f"Approximation of integral: {riemann_sum}")
```
