```python
import numpy as np
import matplotlib.pyplot as plt

# Define the original function f(x) and its antiderivative F(x)
def f(x):
    return x**2  # The original function

def F(x):
    return x**3 / 3  # The antiderivative

# Define the interval and points for the graph
x = np.linspace(0, 3, 500)  # Interval [0, 3]
x_accum = np.linspace(0, 3, 100)  # Points for accumulation

# Compute values for accumulation
F_values = F(x_accum)

# Plot the original function and the accumulation function
plt.figure(figsize=(10, 6))

# Plot f(x)
plt.subplot(2, 1, 1)
plt.plot(x, f(x), label=r"$f(x) = x^2$", color="blue")
plt.title("Original Function and Accumulation Function")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(alpha=0.3)

# Plot F(x) = âˆ« f(t) dt
plt.subplot(2, 1, 2)
plt.plot(x_accum, F_values, label=r"$F(x) = \int_0^x t^2 dt$", color="green")
plt.fill_between(x_accum, 0, F_values, color="green", alpha=0.2, label="Accumulated Area")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel("x")
plt.ylabel("Accumulation: F(x)")
plt.legend()
plt.grid(alpha=0.3)

# Show the plot
plt.tight_layout()
plt.show()
```

# Finding antiderivative of a function

![](./images/img_7.png)

```python
# Importing necessary libraries
import sympy as sp

# Define the variable and the function
x = sp.Symbol('x')
function = 3*x**2 + 4*x - 5

# Finding the antiderivative (indefinite integral)
antiderivative = sp.integrate(function, x)
print(f"Antiderivative of the function: {antiderivative}")

# Evaluating the definite integral from x=1 to x=3
definite_integral = sp.integrate(function, (x, 1, 3))
print(f"Definite integral of the function from x=1 to x=3: {definite_integral}")

# Illustrating the function and the area under the curve using matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Define the function as a lambda for numerical operations
f = lambda x: 3*x**2 + 4*x - 5

# Define the range for plotting
x_vals = np.linspace(0, 4, 100)
y_vals = f(x_vals)

# Plot the function
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='$f(x) = 3x^2 + 4x - 5$', color='blue')

# Fill the area under the curve between x=1 and x=3
x_fill = np.linspace(1, 3, 100)
y_fill = f(x_fill)
plt.fill_between(x_fill, y_fill, color='lightblue', alpha=0.5, label='Area under $f(x)$ from 1 to 3')

# Add labels, legend, and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function and Area Under the Curve')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```
