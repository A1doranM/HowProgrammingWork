```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2

def f_prime(x):
    return 2*x

# Interval [a, b]
a, b = 1, 3

# Compute secant slope
secant_slope = (f(b) - f(a)) / (b - a)

# Find the c value where f'(c) = secant_slope
c = secant_slope / 2  # Solve 2c = secant_slope

# Plot the function
x = np.linspace(0, 4, 500)
plt.plot(x, f(x), label='f(x) = x^2', color='blue')

# Plot the secant line
plt.plot([a, b], [f(a), f(b)], label='Secant Line', linestyle='--', color='orange')

# Plot the tangent line at c
tangent_line = secant_slope * (x - c) + f(c)
plt.plot(x, tangent_line, label=f'Tangent Line at c = {c:.2f}', linestyle='--', color='green')

# Highlight points a, b, and c
plt.scatter([a, b, c], [f(a), f(b), f(c)], color='red', zorder=5)
plt.text(a, f(a), f"A({a}, {f(a)})", fontsize=10)
plt.text(b, f(b), f"B({b}, {f(b)})", fontsize=10)
plt.text(c, f(c), f"C({c:.2f}, {f(c):.2f})", fontsize=10)

# Add labels and legend
plt.title('Mean Value Theorem')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(alpha=0.4)
plt.show()
```
