# Horizontal tangent to implicit curve

Write the equation of the horizontal line that is tangent to the curve and is above the x-axis

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the implicit equation of the curve
def curve(x, y):
    return x**2 + y**4 + 6*x - 7

# Generate x and y values
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = curve(X, Y)

# Plot the curve
plt.contour(X, Y, Z, levels=[0], colors='blue')
plt.axhline(y=2, color='red', linestyle='--', label='y = 2 (Horizontal Tangent)')
plt.title("Curve and Horizontal Tangent Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
