```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^3 - 6x + 2
def f(x):
    return x**3 - 6*x + 2

# Generate x values from 0 to 4
x_values = np.linspace(0, 4, 400)
y_values = f(x_values)

# Plot the function
plt.plot(x_values, y_values, label=r'$f(x) = x^3 - 6x + 2$')

# Plot the x-axis
plt.axhline(0, color='black', linewidth=1)

# Plot points at x = 1 and x = 3
plt.scatter([1, 3], [f(1), f(3)], color='red', zorder=5)
plt.text(1, f(1), f'f(1) = {f(1)}', color='red', fontsize=10, ha='right')
plt.text(3, f(3), f'f(3) = {f(3)}', color='red', fontsize=10, ha='left')

# Label the graph
plt.title(r'Intermediate Value Theorem for $f(x) = x^3 - 6x + 2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```
