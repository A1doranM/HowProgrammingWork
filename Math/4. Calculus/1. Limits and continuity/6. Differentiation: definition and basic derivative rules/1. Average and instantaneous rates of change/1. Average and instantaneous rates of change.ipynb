```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the points for the secant line
x1, x2 = 1, 3
y1, y2 = f(x1), f(x2)

# Generate x values for the function
x_values = np.linspace(0, 4, 400)
y_values = f(x_values)

# Slope of the secant line
secant_slope = (y2 - y1) / (x2 - x1)

# Equation of the secant line: y = slope * (x - x1) + y1
secant_line = secant_slope * (x_values - x1) + y1

# Plot the function and the secant line
plt.plot(x_values, y_values, label=r'$f(x) = x^2$')
plt.plot(x_values, secant_line, '--', label='Secant line', color='orange')

# Plot the points (1, 1) and (3, 9)
plt.scatter([x1, x2], [y1, y2], color='red', zorder=5)
plt.text(x1, y1, f'({x1}, {y1})', color='red', ha='right')
plt.text(x2, y2, f'({x2}, {y2})', color='red', ha='left')

# Label the graph
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(r'Average Rate of Change and Secant Line for $f(x) = x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
