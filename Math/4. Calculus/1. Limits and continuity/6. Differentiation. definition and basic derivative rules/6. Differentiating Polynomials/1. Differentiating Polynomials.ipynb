```python
import numpy as np
import matplotlib.pyplot as plt

# Define the polynomial function f(x) = x^3 - 3x^2 + 4
def f(x):
    return x**3 - 3*x**2 + 4

# Define the derivative of the polynomial f'(x) = 3x^2 - 6x
def f_prime(x):
    return 3*x**2 - 6*x

# Define the tangent line at x = 2
def tangent_line(x):
    return f_prime(2) * (x - 2) + f(2)  # Slope f'(2) and point f(2)

# Generate x values for the plot
x_values = np.linspace(-1, 4, 400)
y_values = f(x_values)

# Generate y values for the tangent line
tangent_values = tangent_line(x_values)

# Plot the polynomial function
plt.plot(x_values, y_values, label=r'$f(x) = x^3 - 3x^2 + 4$')

# Plot the tangent line at x = 2
plt.plot(x_values, tangent_values, '--', label='Tangent at x=2', color='orange')

# Plot the point of tangency (2, 0)
plt.scatter([2], [f(2)], color='red', zorder=5)
plt.text(2, f(2), '(2, 0)', color='red', fontsize=12, ha='left')

# Label the graph
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title(r'Tangent Line to $f(x) = x^3 - 3x^2 + 4$ at $x = 2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
