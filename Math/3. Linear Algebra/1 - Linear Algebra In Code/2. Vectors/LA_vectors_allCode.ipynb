# Unit vector
---

Unit vectors are vectors whose magnitude is exactly 1 unit. They are very useful for different reasons. Specifically, the unit vectors i = [0,1] and j = [1,0] can form together any other vector. 

So for example if we want to create vector V which goes 3 units left and 5 units to top we can write it as V = 3i + 5j or V = [3i, 5j]


### How to find unit vector 

**Step 1: Write the given vector and note its component in x, y, and z directions respectively.**

**Step 2: Find the magnitude (length of the vector) of the vector using the formula,**

In general it is: |A| = (a^2 + b^2 + c^2)

 - For a vector V with endpoints at (x1, y1) and (x2, y2), its magnitude is: |v| = ((x2 - x1)^2 + (y2 - y1)^2)

 - For a vector v when one of its endpoints is at origin (0,0) and the other endpoint is at (x, y), its magnitude is: |v| = (x2 + y2)

**Step 3: Find the unit vector by using the formula,**

Unit Vector = Vector / Magnitude of Vector

```python
# vector
v1 = np.array([-3, 6])

# unit vector
# Step 1: Write the given vector and note its component 
# [-3, 6]

# Step 2: Find the magnitude
v1magnitude = np.sqrt(pow(v1[0], 2) + pow(v1[1], 2))

# Step 3: Find the unit vector 
v1unit = np.array([v1[0] / v1magnitude, v1[1] / v1magnitude])

# plot them
plt.plot([0, v1unit[0]], [0, v1unit[1]], 'r', label='v1-unit', linewidth=5)
plt.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')

# axis square
plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.grid()
plt.legend()
plt.show()
```

# Linear combination and Span
---

The Span of vector, or two vectors is just the all possible variants of vectors which we can create when multiplying, dividing, original vector, or vectors. The formula will look like this, cA1 + cA2 + ... cAn, where are c is any real number (R).  


For example, we have two vectors a = [1, 2] and b = [0, 3]

1) Lets multiply a by 0 and b by 0, 0a + 0b = [0, 0],
2) Another case, multiply a by 3 and b by -2, 3a - 2b = [3-0, 6-6] = [3, 0]
3) As we can see the span (or the whole number of vectors variants) of these two vectors are R^2 

Let's take another example where two vectors are on the same line. A = [2, 2], B = [-2, -2], the Span of these vectors will be the line where they lie, because whenever linear combination we apply to them, we are never obtain vector C, or another vector outside the line. 

![img](./images/img_1.png)

```python
# set S
S1 = np.array([1, 1, 0])
S2 = np.array([1, 7, 0])

# vectors v and w
v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

# draw vectors
fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
ax = plt.axes(projection='3d')
ax.plot([0, S1[0]], [0, S1[1]], [.1, S1[2] + .1], 'r', linewidth=3)
ax.plot([0, S2[0]], [0, S2[1]], [.1, S2[2] + .1], 'r', linewidth=3)

ax.plot([0, v[0]], [0, v[1]], [.1, v[2] + .1], 'g', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'y')

# now draw plane
xx, yy = np.meshgrid(range(-15, 16), range(-15, 16))
cp = np.cross(S1, S2)
z1 = (-cp[0] * xx - cp[1] * yy) * 1. / cp[2]
ax.plot_surface(xx, yy, z1)

plt.show()
```

# Linear independence and Span
---

Linear dependence means that we can represent one vector from another, e.g. A = [2, 3], B = [2, 6] as we can see we can represent vector 
B as 2A, or vector A as B/2. The Span of this two vectors will be the line because they can represent only itself. 

Linear independence means that we can not represent one vector from another, e.g. A = [0, 3], B = [4, 6], these vectors are linearly independent because we can not represent A using B and vice versa. And the span of this vectors will be R^2.   

To determine if set of vectors are linearly independent we should solve system of linear equations and the only possible solution for this 
system should be [0, 0, ....., 0]

# Subspaces and the basis for a subspace
---


Subspace of some space is the area of vectors which contains zero vector, and meets some requirements:  

![subspace](./images/img_6.png)

1) Closure under scalar multiplication, as we see on image some if some vector X are in subspace if we multiply it on scalar (CX)
    it should remain in this subspace
2) Closure under addition, if some vectors are in subspace their sum should be also in subspace. 

    ![subspace](./images/img_7.png)


Let's see some example. On image below we want to check if S(with X >= 0) is the subspace of R^2, as we see it contains zero vector,

### Basis of the subspace 

The Basis of the subspace is the **minimum linearly independent set of vectors that Span the subspace**, there are may be more than one basis for subspace.

# Vector-vector multiplication: the dot product
---


The dot vector product is the scalar which can be calculated in two ways 

![dot_product](./images/img_9.png)

If we think why we use cos, here is the answer 

![dot_product](./images/img_10.png)

```python
## many ways to compute the dot product

v1 = np.array([1, 2, 3, 4, 5])
v2 = np.array([0, -4, -3, 6, 5])

# method 1
dp1 = sum(np.multiply(v1, v2))

# method 2
dp2 = np.dot(v1, v2)

# method 3
dp3 = np.matmul(v1, v2)

# method 4
dp4 = 0  # initialize

# loop over elements
for i in range(len(v1)):
    # multiply corresponding element and sum
    dp4 = dp4 + v1[i] * v2[i]

print(dp1, dp2, dp3, dp4)
```

# Dot product properties: associative and distributive

```python
## Distributive property

# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot(a, (b + c))
res2 = np.dot(a, b) + np.dot(a, c)

# compare them
print([res1, res2])
```

```python
## Associative property

# create random vectors
n = 5
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# the two results
res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)

# compare them
print(res1)
print(res2)


### special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

# The dot product from a geometric perspective
---


```python

# two vectors
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])

# compute the angle (radians) between two vectors
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

# draw them
fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
ax = plt.axes(projection='3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' % ang)
plt.show()
```

```python
## equivalence of algebraic and geometric dot product formulas

# two vectors
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3])

# algebraic
dp_a = np.dot(v1, v2)

# geometric
dp_g = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(ang)

# print dot product to command
print(dp_a)
print(dp_g)
```

# Vector cross product
---

Cross product can be defined only in R3, it gives a vector as an answer, and is sometimes called the vector product. The Cross Product a  b of two vectors is another vector that is at right angles to both:

![cross_product](./images/img_14.png)

![cross_product](./images/img_15.png)

We can calculate cross product in two ways 

First way: 
    
    a  b = |a| |b| sin() n
    
    |a| is the magnitude (length) of vector a
    |b| is the magnitude (length) of vector b
     is the angle between a and b 
    n is the unit vector at right angles to both a and b
    
![cross_product](./images/img_16.png)
    
Second way: 

    When a and b start at the origin point (0,0,0), the Cross Product will end at:

    cx = AyBz  AzBy
    cy = AzBx  AxBz
    cz = AxBy  AyBx

```python
# create vectors
v1 = [-3, 2, 5]
v2 = [4, -3, 0]

# Python's cross-product function
v3a = np.cross(v1, v2)

# "manual" method
v3b = [[v1[1] * v2[2] - v1[2] * v2[1]],
       [v1[2] * v2[0] - v1[0] * v2[2]],
       [v1[0] * v2[1] - v1[1] * v2[0]]]

print(v3a, v3b)

fig = plt.figure()
# ax = fig.add_subplot(projection='3d')
ax = plt.axes(projection='3d')

# draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10, 10, 10))
z1 = (-v3a[0] * xx - v3a[1] * yy) / v3a[2]
ax.plot_surface(xx, yy, z1, alpha=.2)

## plot the two vectors
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'r')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'g')
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'b')

ax.view_init(azim=150, elev=45)
plt.show()
```

```python

v1 = np.array([1, 2, 3])
v2 = np.array([-1, 0, 1])

# outer product
np.outer(v1, v2)

# terrible programming, but helps conceptually:
op = np.zeros((len(v1), len(v2)))
for i in range(0, len(v1)):
    for j in range(0, len(v2)):
        op[i, j] = v1[i] * v2[j]

print(op)
```

# Hermitian transpose (a.k.a. conjugate transpose)
---


```python
# create a complex number
z = complex(3, 4)

# magnitude
print(np.linalg.norm(z))

# by transpose?
print(np.transpose(z) * z)

# by Hermitian transpose
print(np.transpose(z.conjugate()) * z)

# complex vector
v = np.array([3, 4j, 5 + 2j, complex(2, -5)])
print(v.T)
print(np.transpose(v))
print(np.transpose(v.conjugate()))
```
