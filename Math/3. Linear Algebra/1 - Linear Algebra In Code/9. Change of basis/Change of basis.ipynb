```python
import numpy as np
import matplotlib.pyplot as plt

# Define the old basis vectors
b1 = np.array([1, 0])
b2 = np.array([0, 1])

# Define the new basis vectors
c1 = np.array([2, 1])
c2 = np.array([-1, 1])

# Define the vector in the old basis
v = np.array([3, 2])

# Change of basis matrix
P = np.array([c1, c2]).T

# Find the inverse of the change of basis matrix
P_inv = np.linalg.inv(P)

# Compute the new coordinates of the vector in the new basis
v_new_basis = P_inv @ v

# Plotting
plt.figure()
plt.quiver(0, 0, b1[0], b1[1], angles='xy', scale_units='xy', scale=1, color='r', label='b1')
plt.quiver(0, 0, b2[0], b2[1], angles='xy', scale_units='xy', scale=1, color='g', label='b2')
plt.quiver(0, 0, c1[0], c1[1], angles='xy', scale_units='xy', scale=1, color='b', label='c1')
plt.quiver(0, 0, c2[0], c2[1], angles='xy', scale_units='xy', scale=1, color='purple', label='c2')
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='orange', label='v in B')
plt.quiver(0, 0, v_new_basis[0], v_new_basis[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='v in C')

plt.xlim(-3, 4)
plt.ylim(-3, 4)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Change of Basis Visualization')
plt.show()
```

## Find the transformation matrix for an Alternate Basis 

![](./images/img_9.png)

![](./images/img_12.png)

```python
import numpy as np

# Original transformation matrix A
A = np.array([[2, 1], [1, 3]])

# New basis vectors
b1 = np.array([1, 1])
b2 = np.array([-1, 2])

# Change of basis matrix P
P = np.array([b1, b2]).T

# Inverse of P
P_inv = np.linalg.inv(P)

# Transformation matrix in the new basis B
# The @ symbol in Python is used for matrix multiplication, introduced in Python 3.5 as part of PEP 465.
# Equivalent Without "@" is A_B = np.dot(np.dot(P_inv, A), P)
A_B = P_inv @ A @ P

print("Transformation matrix in the new basis B:")
print(A_B)
```
