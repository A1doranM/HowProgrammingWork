## Visualization using Python
We'll visualize the concept of orthogonal complements in 2D and 3D using Python, NumPy, and Matplotlib.

![](./images/img_1.png)

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a vector in 2D
vector_w = np.array([2, 1])

# Define the orthogonal complement (perpendicular vector)
vector_w_perp = np.array([-1, 2])

# Plotting
plt.figure()
plt.quiver(0, 0, vector_w[0], vector_w[1], angles='xy', scale_units='xy', scale=1, color='r', label='W')
plt.quiver(0, 0, vector_w_perp[0], vector_w_perp[1], angles='xy', scale_units='xy', scale=1, color='b', label='W⊥')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Orthogonal Complements in 2D')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define vectors in 3D
vector_w1 = np.array([1, 2, 0])
vector_w2 = np.array([2, -1, 1])

# Compute a vector in the orthogonal complement using the cross product
vector_w_perp = np.cross(vector_w1, vector_w2)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, vector_w1[0], vector_w1[1], vector_w1[2], color='r', label='W1')
ax.quiver(0, 0, 0, vector_w2[0], vector_w2[1], vector_w2[2], color='g', label='W2')
ax.quiver(0, 0, 0, vector_w_perp[0], vector_w_perp[1], vector_w_perp[2], color='b', label='W⊥')
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Orthogonal Complements in 3D')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Define the vectors in the subspace W
vector_w1 = np.array([1, 2, 3])
vector_w2 = np.array([4, 5, 6])
# Define the vector in the orthogonal complement W⊥
vector_w_perp = np.array([-1, 2, -1])
# Plotting the subspace W
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, vector_w1[0], vector_w1[1], vector_w1[2], color='r', label='W1')
ax.quiver(0, 0, 0, vector_w2[0], vector_w2[1], vector_w2[2], color='g', label='W2')
# Plotting the orthogonal complement W⊥
ax.quiver(0, 0, 0, vector_w_perp[0], vector_w_perp[1], vector_w_perp[2], color='b', label='W⊥')
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Orthogonal Complements in 3D')
ax.legend()
plt.show()
```
