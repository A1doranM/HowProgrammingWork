# Exploring One-Variable Quantitative Data

When we have a set of numerical data for a single variable (like the heights of students in a class, or the prices of houses in a neighborhood), our goal is to understand its **distribution**. We want to answer questions like:

*   What is a typical value? (Center)
*   How spread out are the values? (Spread)
*   What is the overall shape of the data? (Shape)

Percentiles, z-scores, and the normal distribution are powerful tools that help us describe the position of individual data points and understand the overall shape of the data.

---

## 1. Percentiles

### What is a Percentile?

A percentile is a measure that tells us what percentage of the total data falls *below* a certain value.

For example, if you score in the **85th percentile** on a standardized test, it means you scored higher than 85% of the people who took the test. It's a way to understand a value's rank or position relative to the entire dataset.

**Key Percentiles:**
*   **The 50th Percentile:** This is the **median** of the data. 50% of the data is below this value, and 50% is above it.
*   **The 25th Percentile (Q1):** Also known as the first quartile. 25% of the data is below this value.
*   **The 75th Percentile (Q3):** Also known as the third quartile. 75% of the data is below this value.

### How to Calculate a Percentile (A Simple Method)

1.  **Order the Data:** Arrange your dataset of `n` values in ascending order (from smallest to largest).
2.  **Calculate the Rank:** To find the rank `R` for a given percentile `P`, use the formula:
    `R = (P / 100) * (n + 1)`
3.  **Find the Value:**
    *   If `R` is a whole number, the percentile value is the data point at the `R`-th position in your ordered list.
    *   If `R` is a decimal, you need to interpolate. Let `I` be the integer part and `F` be the fractional part of `R`. The value is found by:
      `Value(I) + F * (Value(I+1) - Value(I))`

**Example Calculation:**
Let's find the 70th percentile for the dataset: `[10, 50, 20, 40, 30]`

1.  **Order Data:** `[10, 20, 30, 40, 50]`. Here, `n = 5`.
2.  **Calculate Rank:** `R = (70 / 100) * (5 + 1) = 0.70 * 6 = 4.2`.
3.  **Find Value:** `R` is not a whole number.
    *   The integer part `I` is 4. The 4th value is `40`.
    *   The fractional part `F` is 0.2.
    *   The next value (5th) is `50`.
    *   Interpolate: `40 + 0.2 * (50 - 40) = 40 + 0.2 * 10 = 40 + 2 = 42`.
    *   So, the **70th percentile is 42**.

### Real-Life Uses

*   **Standardized Test Scores (SAT, GRE):** Your score report always shows your percentile rank to help you understand your performance relative to others.
*   **Pediatric Growth Charts:** Doctors use percentiles to track a child's height and weight for their age, identifying if they are growing as expected. A child in the 95th percentile for height is taller than 95% of children of the same age.
*   **Website Performance:** Companies track the 95th or 99th percentile of page load times. This tells them the worst-case experience for most users, which is often a more important metric than the average load time.

## 2. Z-Scores

### What is a Z-Score?

A **z-score** (or standard score) measures exactly how many **standard deviations** a data point is from the **mean** of the dataset. It's a way of *standardizing* values.

*   A **positive** z-score means the data point is **above** the mean.
*   A **negative** z-score means the data point is **below** the mean.
*   A z-score of **0** means the data point is **exactly the mean**.

The key benefit is that you can compare values from different datasets (e.g., your score on a history test vs. your score on a math test) by putting them on a common scale.

### How to Calculate a Z-Score

The formula is:

`z = (x - μ) / σ`

Where:
*   `x` is the individual data point.
*   `μ` (mu) is the mean of the dataset.
*   `σ` (sigma) is the standard deviation of the dataset.

**Example Calculation:**
A student scores **90** on a test. The class mean (`μ`) is **75**, and the standard deviation (`σ`) is **5**.

1.  **Calculate z-score:** `z = (90 - 75) / 5 = 15 / 5 = 3`.
2.  **Interpretation:** The student's score is **3 standard deviations above** the class average. This is an exceptionally good score.

### Real-Life Uses

*   **Comparing Different Scales:** An admissions officer wants to compare two students. One scored 1400 on the SAT (mean=1050, std=200) and the other scored 32 on the ACT (mean=21, std=5).
    *   SAT z-score: `(1400 - 1050) / 200 = 1.75`
    *   ACT z-score: `(32 - 21) / 5 = 2.2`
    *   The student with the ACT score of 32 performed better relative to their peers.
*   **Outlier Detection:** In data analysis, a data point with a z-score of greater than +3 or less than -3 is often considered an **outlier**, as it is very far from the mean.
*   **Quality Control:** A factory produces bolts with a target diameter. By calculating the z-score for each bolt's diameter, they can quickly identify and discard any that deviate too much from the standard.

## 3. The Normal Distribution

### What is the Normal Distribution?

The **normal distribution** is a specific type of data distribution that is perfectly **symmetric** and **bell-shaped**. Many natural phenomena, such as human height, blood pressure, and measurement errors, tend to follow this pattern.

**Key Characteristics:**
*   The curve is symmetric around the mean.
*   The **mean, median, and mode are all equal** and located at the center.
*   It's defined by its mean (`μ`) and standard deviation (`σ`).

The most important feature for practical use is the **Empirical Rule (or 68-95-99.7 Rule)**.

![Empirical Rule](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/600px-Standard_deviation_diagram.svg.png)

This rule states that for a normal distribution:
*   Approximately **68%** of the data falls within **1 standard deviation** of the mean (`μ ± 1σ`).
*   Approximately **95%** of the data falls within **2 standard deviations** of the mean (`μ ± 2σ`).
*   Approximately **99.7%** of the data falls within **3 standard deviations** of the mean (`μ ± 3σ`).

### Real-Life Uses

*   **Statistical Inference:** The normal distribution is the foundation for many statistical tests. The Central Limit Theorem states that if you take large enough samples from any population, the distribution of the sample means will be approximately normal.
*   **Financial Modeling:** Stock returns and other financial metrics are often assumed to be normally distributed to model risk and predict the probability of certain price movements.
*   **Predicting Probabilities:** If we know that IQ scores are normally distributed with a mean of 100 and a standard deviation of 15, we can use the Empirical Rule to predict that about 95% of the population has an IQ between 70 (`100 - 2*15`) and 130 (`100 + 2*15`). We can use z-scores to find the probability for any score, not just these clean cutoffs.

## Python Code Illustration

This code will generate a normally distributed dataset of test scores and then demonstrate how to calculate percentiles and z-scores, and finally visualize the distribution with the Empirical Rule.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- 1. SETUP: Create a sample dataset ---
# Let's generate 1000 test scores that are normally distributed.
# Mean score (μ) = 80
# Standard deviation (σ) = 10
mean_score = 80
std_dev = 10
num_scores = 1000

# Set a seed for reproducibility so we get the same random numbers every time
np.random.seed(42)
test_scores = np.random.normal(loc=mean_score, scale=std_dev, size=num_scores)

print("--- Sample Data ---")
print(f"First 5 scores: {test_scores[:5].round(2)}")
print(f"Mean of generated scores: {np.mean(test_scores):.2f}")
print(f"Standard deviation of generated scores: {np.std(test_scores):.2f}\n")


# --- 2. Exploring with PERCENTILES ---
print("--- Percentile Calculations ---")

# Calculate the 25th, 50th (median), and 75th percentiles
p25 = np.percentile(test_scores, 25)
p50 = np.percentile(test_scores, 50)
p75 = np.percentile(test_scores, 75)
p90 = np.percentile(test_scores, 90)

print(f"25th Percentile (Q1): A score of {p25:.2f} is higher than 25% of the class.")
print(f"50th Percentile (Median): A score of {p50:.2f} is the middle score.")
print(f"75th Percentile (Q3): A score of {p75:.2f} is higher than 75% of the class.")
print(f"90th Percentile: A score of {p90:.2f} is required to be in the top 10%.\n")

# We can also do the reverse: what percentile is a specific score?
# Let's check a score of 95.
score_to_check = 95
percentile_of_score = stats.percentileofscore(test_scores, score_to_check)
print(f"A score of {score_to_check} is in the {percentile_of_score:.2f}th percentile.\n")


# --- 3. Exploring with Z-SCORES ---
print("--- Z-Score Calculations ---")

# Let's find the z-score for a student who scored 95
student_score = 95
z_score = (student_score - np.mean(test_scores)) / np.std(test_scores)
print(f"The z-score for a student who scored {student_score} is: {z_score:.2f}")
print(f"This means the score is {z_score:.2f} standard deviations above the mean.\n")

# Let's find the z-score for a student who scored 62
student_score_2 = 62
z_score_2 = (student_score_2 - np.mean(test_scores)) / np.std(test_scores)
print(f"The z-score for a student who scored {student_score_2} is: {z_score_2:.2f}")
print(f"This means the score is {abs(z_score_2):.2f} standard deviations below the mean.\n")


# --- 4. Visualizing the NORMAL DISTRIBUTION and the Empirical Rule ---
print("--- Visualizing the Distribution ---")
plt.figure(figsize=(12, 6))
sns.histplot(test_scores, kde=True, bins=30, color='skyblue')

# Add vertical lines for the mean and standard deviations
plt.axvline(mean_score, color='red', linestyle='--', linewidth=2, label=f'Mean ({mean_score:.2f})')
plt.axvline(mean_score + std_dev, color='green', linestyle='--', linewidth=2, label=f'1 Std Dev ({mean_score + std_dev:.2f})')
plt.axvline(mean_score - std_dev, color='green', linestyle='--', linewidth=2)
plt.axvline(mean_score + 2 * std_dev, color='purple', linestyle='--', linewidth=2, label=f'2 Std Dev ({mean_score + 2*std_dev:.2f})')
plt.axvline(mean_score - 2 * std_dev, color='purple', linestyle='--', linewidth=2)

# Add text for the Empirical Rule
plt.text(mean_score, 5, '68%', fontsize=12, ha='center', color='black')
plt.text(mean_score - 1.5 * std_dev, 20, '~95%', fontsize=12, ha='center', color='black')
plt.text(mean_score + 1.5 * std_dev, 20, '~95%', fontsize=12, ha='center', color='black')

# Final plot touches
plt.title('Distribution of Test Scores with Empirical Rule', fontsize=16)
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
