## Get prices for General Electric (GE)

```python
import pandas as pd
import yfinance as yf
pd.options.display.float_format = '{:.4f}'.format
```

```python
start="2015-10-22"
end="2025-10-22"
symbol=["GE"]
```

```python
df = yf.download(symbol, start, end)
```

```python
df
```

### Select "Close" prices only and create a price chart for GE

```python
close = df["Close"].copy()
close
```

```python
close["GE"].dropna().plot(figsize=(15, 8), fontsize=13, logy=True)
```
