# Price Changes and Financial Returns

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
close =  pd.read_csv("close.csv", index_col="Date", parse_dates=["Date"])
close
```

```python
ba = close.BA.dropna().to_frame().copy() # Get Boeing Close prices and make pandas DataFrame from it.
```

```python
ba
```

```python
ba.rename(columns={"BA": "Price"}, inplace=True)
```

```python
ba
```

```python
ba["P_lag1"] = ba["Price"].shift(periods=1) # Shift specified column values forward on 1 day, so day 2 became 3, 3 is 4, and so on.
```

```python
ba
```

```python
# Now we can calculate absolute price changes 

ba["P_diff"] = ba.Price.sub(ba.P_lag1)
ba
```

```python
# Another way to calculate the same 
ba["P_diff2"] = ba.Price.diff(periods=1)
```

```python
ba
```

# Absolute Price Changes is not meaningful 

Because One company can cost 500$ and price change in 10$ is relatively small but for another company it can be large value

## Lets Calculate Relative price changes instead

```python
ba.Price.div(ba.P_lag1) # Alternative 1 
```

```python
ba["Returns"] = ba.Price.pct_change(periods=1) # Alternative 2
ba
```

```python
ba.drop(columns=["P_lag1", "P_diff", "P_diff2"], inplace=True)
```

```python
ba
```
